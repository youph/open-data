This script builds a classifier (a Stacked Denoising Autoencoder (SdA), a logistic regression model, a random forest, and a clustering classifier), and trains it on the dataset of raw "bag of character" (or other) representations of samples drawn from a set of labeled data columns with known and unknown attribute types. 

**Note on the terms:**

- schemas: characterizes different general standards used by bodies who own data files. E.g., data files from council 1 all adhere to the same schema, while data files from council 2 all adhere to a different schema.
- data files: raw data files (in csv format) with some data tables. We split the data files into training and testing sets of data files, according to the schemas they belong to.
- instances (buckets): columns in data files: training data files give us the training instances/buckets that are then divided into non-overlapping training and validation sets of buckets (thus training and validation bucket sets are from the same collection of "known" data files), and testing data files give us the testing instances/buckets (representing columns from "new", previously unseen data files). Buckets are labeled with one of the "known" attribute class labels, or unlabeled (labeled with "UNKNOWN" attribute class label). 
- examples: points in the feature space, drawn from the buckets. Examples are labeled with the labels of their source buckets. Training examples are drawn from training buckets, validation examples - from validation buckets, testing examples - from testing buckets. Examples are drawn from each bucket separately, i.e., buckets of the same class are NOT consolidated when generating examples.
- examples_per_bucket: number of examples drawn from each bucket
- examples_per_attr (OBSOLETE): total number of examples per attribute class that should be drawn in the training, validation, and testing set. OBSOLETE, replaced with examples_per_bucket, the example sets are now not being balanced on the number of examples per class.
- records_per_example: number of data records (rows) used for generating a single example. Examples are generated by taking a random sample of 'records_per_example' records (rows) from the bucket, concatenating them into a single string, and extracting the feature vector from this string (e.g., character frequencies).

```python
def logistic_model(datasets, learning_rate=0.13, n_epochs=1000,
                           batch_size=10, train_size=None, valid_size=None, test_size=None):
    """
    Demonstrate stochastic gradient descent optimization of a log-linear
    model

    :type learning_rate: float
    :param learning_rate: learning rate used (factor for the stochastic
                          gradient)

    :type n_epochs: int
    :param n_epochs: maximal number of epochs to run the optimizer

    :type dataset: string

    """

    train_set_x, train_set_y = datasets[0]
    valid_set_x, valid_set_y = datasets[1]
    test_set_x, test_set_y = datasets[2]
    
    print ('.... shape of train_set_x:')
    print (train_set_x.get_value(borrow=True).shape)
    print ('.... shape of valid_set_x:')
    print (valid_set_x.get_value(borrow=True).shape)
    print ('.... shape of test_set_x:')
    print (test_set_x.get_value(borrow=True).shape)
    
    # compute number of minibatches for training, validation and testing
    n_train_batches = train_set_x.get_value(borrow=True).shape[0] / batch_size
    n_valid_batches = valid_set_x.get_value(borrow=True).shape[0] / batch_size
    n_test_batches = test_set_x.get_value(borrow=True).shape[0] / batch_size

    ######################
    # BUILD ACTUAL MODEL #
    ######################
    print ('... building the model')

    # allocate symbolic variables for the data
    index = T.lscalar()  # index to a [mini]batch

    # generate symbolic variables for input (x and y represent a
    # minibatch)
    x = T.matrix('x')  # data, presented as rasterized images
    y = T.ivector('y')  # labels, presented as 1D vector of [int] labels

    # construct the logistic regression class
    # This is a 15-class classification problem, hence n_out=len(attributes)=15
    classifier = LogisticRegression(input=x, n_in=train_set_x.get_value(borrow=True).shape[1], n_out=len(attributes))

    # the cost we minimize during training is the negative log likelihood of
    # the model in symbolic format
    cost = classifier.negative_log_likelihood(y)

    # compiling a Theano function that computes the mistakes that are made by
    # the model on a minibatch
    test_model = theano.function(
        inputs=[index],
        outputs=classifier.errors(y),
        givens={
            x: test_set_x[index * batch_size: (index + 1) * batch_size],
            y: test_set_y[index * batch_size: (index + 1) * batch_size]
        }
    )

    validate_model = theano.function(
        inputs=[index],
        outputs=classifier.errors(y),
        givens={
            x: valid_set_x[index * batch_size: (index + 1) * batch_size],
            y: valid_set_y[index * batch_size: (index + 1) * batch_size]
        }
    )

    # compute the gradient of cost with respect to theta = (W,b)
    g_W = T.grad(cost=cost, wrt=classifier.W)    # g_W and g_b are symbolic variables.
    g_b = T.grad(cost=cost, wrt=classifier.b)

    # start-snippet-3
    # specify how to update the parameters of the model as a list of
    # (variable, update expression) pairs. In each pair, the first element is the symbolic variable to be updated in the step,
    # and the second element is the symbolic function for calculating its new value.
    updates = [(classifier.W, classifier.W - learning_rate * g_W),
               (classifier.b, classifier.b - learning_rate * g_b)]

    # compiling a Theano function `train_model` that returns the cost, but in
    # the same time updates the parameter of the model based on the rules
    # defined in `updates`. `train_model` performs one step of gradient descent.
    train_model = theano.function(
        inputs=[index],
        outputs=cost,
        updates=updates,
        givens={
            x: train_set_x[index * batch_size: (index + 1) * batch_size],
            y: train_set_y[index * batch_size: (index + 1) * batch_size]
        }  # givens is a dictionary (key: value) whose keys are symbolic variables (x and y) and whose values specify their replacements during the step.
    )
    # end-snippet-3

    ###############
    # TRAIN MODEL #
    ###############
    print ('... training the model (for maximum of %i epochs, with %i minibatches per epoch)' % (n_epochs, n_train_batches))
    # early-stopping parameters
    patience = patience = max(10000, 10 * n_train_batches)  # look as this many examples regardless
    patience_increase = 2.  # wait this much longer when a new best is found
        
    print ('... patience = %i' %patience)
    improvement_threshold = 0.995  # a relative improvement of this much is
                                  # considered significant
    validation_frequency = min(n_train_batches, patience / 2)
                                  # go through this many
                                  # minibatches before checking the network
                                  # on the validation set; in this case we
                                  # check every epoch

    best_validation_loss = numpy.inf
    test_score = 0.
    start_time = timeit.default_timer()

    # Each time train_model(index) is called, it will compute and return the cost of a minibatch, while also performing a step of MSGD. The entire learning algorithm thus consists in looping over all examples in the dataset, considering all the examples in one minibatch at a time, and repeatedly calling the train_model function.
    done_looping = False
    epoch = 0
    while (epoch < n_epochs) and (not done_looping):
        epoch = epoch + 1
        for minibatch_index in xrange(n_train_batches):

            minibatch_avg_cost = train_model(minibatch_index)
            # iteration number
            iter = (epoch - 1) * n_train_batches + minibatch_index

            if (iter + 1) % validation_frequency == 0:
                # compute zero-one loss on validation set
                validation_losses = [validate_model(i)
                                     for i in xrange(n_valid_batches)]  # zero-one losses on minibatches from the validation set
                this_validation_loss = numpy.mean(validation_losses)    # average zero-one loss over the entire validation set

                print(
                    'epoch %i, minibatch %i/%i, validation error %f %%, remaining patience %i' %
                    (
                        epoch,
                        minibatch_index + 1,
                        n_train_batches,
                        this_validation_loss * 100.,
                        patience - iter
                    )
                )

                # if we got the best validation score until now
                if this_validation_loss < best_validation_loss:
                    #improve patience if loss improvement is good enough
                    if this_validation_loss < best_validation_loss * improvement_threshold:
                        patience = max(patience, iter * patience_increase)

                    best_validation_loss = this_validation_loss
                    # test it on the test set (why?)
                    test_losses = [test_model(i) for i in xrange(n_test_batches)]
                    test_score = numpy.mean(test_losses)

                    print(
                        (
                            '     epoch %i, minibatch %i/%i, test error of'
                            ' best model %f %%'
                        ) %
                        (
                            epoch,
                            minibatch_index + 1,
                            n_train_batches,
                            test_score * 100.
                        )
                    )

                    # save the best model
                    save_model(model=classifier,hp=hp,file_name='best_logistic_model.pkl')

            if patience <= iter:
                done_looping = True
                break

    end_time = timeit.default_timer()
    print(
        (
            '\nOptimization complete with best validation score of %f %%,  '
            'with test performance %f %%'
        )
        % (best_validation_loss * 100., test_score * 100.)
    )
    print ('The code ran for %d epochs, with %f epochs/sec' % (
        epoch, 1. * epoch / (end_time - start_time)))
    
    print ('The code for logistic model ran for %.1fs' % ((end_time - start_time)))
    
    return (best_validation_loss, test_score)
```

```python
def mlp_model(datasets, hp, numpy_rng=None):
    
    # extract the training, validation, and testing datasets from datasets:
    train_set_x, train_set_y = datasets[0]
    valid_set_x, valid_set_y = datasets[1]
    test_set_x, test_set_y = datasets[2]
    
    # compute number of minibatches for training, validation and testing
    batch_size = hp['batch_size']
    n_train_batches = train_set_x.get_value(borrow=True).shape[0] / batch_size
    n_valid_batches = valid_set_x.get_value(borrow=True).shape[0] / batch_size
    n_test_batches = test_set_x.get_value(borrow=True).shape[0] / batch_size
    
    # Construct the stacked denoising autoencoder class:
    print ('... building the model')
    sda = SdA(
        numpy_rng=numpy_rng,
        n_ins=train_set_x.get_value(borrow=True).shape[1],
        hidden_layers_sizes=hp['hidden_layers_sizes'],
        n_outs=len(attributes), 
        corruption_levels=hp['corruption_levels']
    )
    
    # Pre-train the model:
    print ('... getting the pretraining functions')
    # get the pretraining functions:
    pretraining_fns = sda.pretraining_functions(train_set_x=train_set_x,
                                                batch_size=batch_size)
    # get the training, validation and testing function for the model:
    print ('... getting the finetuning functions')
    train_fn, validate_model, test_model = sda.build_finetune_functions(
                                                                        datasets=datasets,
                                                                        batch_size=batch_size,
                                                                        learning_rate=hp['finetune_lr']
                                                                        )
    
    # Test the freshly initialised model on the test set (should be ~93.3% for the 15-class problem, i.e., pure chance error):
    test_score = 0.
    test_losses = test_model()
    test_score = numpy.mean(test_losses)
    print(('\n Test error of the fresh model: %f %% (should be ~ %f %% for a balanced set of examples)') %(test_score * 100., (1.-1./len(attributes))*100.))
    
    pretraining_epochs = hp['pretraining_epochs']
    pretrain_lr = hp['pretrain_lr']
    print ('\n... pre-training the model for %i epochs per layer, learning rate %f' %(pretraining_epochs,pretrain_lr))
    start_time = timeit.default_timer()
    ## Pre-train layer-wise
    corruption_levels = hp['corruption_levels']
    for i in xrange(sda.n_layers):
        # go through pretraining epochs
        for epoch in xrange(pretraining_epochs):
            # go through the training set
            c = []
            for batch_index in xrange(n_train_batches):
                c.append(pretraining_fns[i](index=batch_index,     #this call of pretraining_fns[i] updates the parameters of the i-th layer of sda?
                             corruption=corruption_levels[i],
                             lr=pretrain_lr))
            print ('Pre-training layer %i, epoch %d, cost ' % (i, epoch)),
            print (numpy.mean(c))

    end_time = timeit.default_timer()

    print ('Pretraining ran for %.2fm' % ((end_time - start_time) / 60.))
    
    # Fine-tune the MLP:
    training_epochs = hp['finetuning_epochs']
    # early-stopping parameters:
    patience = max(10000, 10 * n_train_batches)  # look as this many examples regardless
    patience_increase = 2.  # wait this much longer when a new best is found
    improvement_threshold = 0.995  # a relative improvement of this much is considered significant

    print ('\n... fine-tuning the model with patience = %i' %patience)
    validation_frequency = min(n_train_batches, patience / 2)
                                      # go through this many
                                      # minibatches before checking the network
                                      # on the validation set; in this case we
                                      # check every epoch

    best_validation_loss = numpy.inf
    test_score = 0.
    start_time = timeit.default_timer()
    done_looping = False
    epoch = 0

    while (epoch < training_epochs) and (not done_looping):
        epoch = epoch + 1
        for minibatch_index in xrange(n_train_batches):
            minibatch_avg_cost = train_fn(minibatch_index)
            iter = (epoch - 1) * n_train_batches + minibatch_index

            if (iter + 1) % validation_frequency == 0:
                validation_losses = validate_model()
                this_validation_loss = numpy.mean(validation_losses)
                print('epoch %i, minibatch %i/%i, validation error %f %%, remaining patience %i' %
                          (epoch, minibatch_index + 1, n_train_batches,
                           this_validation_loss * 100., patience-iter))

                # if we got the best validation score until now
                if this_validation_loss < best_validation_loss:
                    #improve patience if loss improvement is good enough
                    if (
                            this_validation_loss < best_validation_loss *
                            improvement_threshold
                        ):
                        patience = max(patience, iter * patience_increase)

                    # save best validation score and iteration number
                    best_validation_loss = this_validation_loss
                    best_iter = iter

                    # test it on the test set
                    test_losses = test_model()
                    test_score = numpy.mean(test_losses)

                    # save the best model found thus far:
                    save_model(model=sda,hp=hp,file_name='best_mlp_model.pkl')

                    print(('     epoch %i, minibatch %i/%i, test error of '
                               'best model %f %%') %
                              (epoch, minibatch_index + 1, n_train_batches,
                               test_score * 100.))

            if patience <= iter:
                done_looping = True
                break

    end_time = timeit.default_timer()
    print(
            (
                '\nOptimization complete after %i iterations with best validation score of %f %%, '
                'on iteration %i, '
                'with test performance %f %%'
            )
            % (iter, best_validation_loss * 100., best_iter + 1, test_score * 100.)
        )
    
    return (best_validation_loss, test_score)
```

```python
def compute_roc_auc(classifier, x, true_y):  # compute per-class tpr, fpr, and ROC AUC, 
                                                        # in this class-vs-all other classes manner, for a given classifier, using data x, 
                                                        # with true labels in true_y 

    true_y_bin = label_binarize(y=true_y, classes=targets_dict.keys())
    
    if str(classifier.__class__) == "<class 'sklearn.ensemble.forest.RandomForestClassifier'>":
        y_score = classifier.predict_proba(x)
        
    if str(classifier.__class__) == "<class 'SdA.SdA'>":
        y_score = mlp_predict(classifier, x, proba=True, verbose=False)
    
    if str(classifier.__class__) == "<class 'logistic_sgd.LogisticRegression'>":
        y_score = logistic_predict(classifier, x, proba=True, verbose=False)
        
    # Compute ROC curve and ROC area for each class, in one-vs-all manner
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    mean_error = dict()   # per-class mean error, in this-vs-other class manner
    for i in targets_dict.keys():
        fpr[i], tpr[i], _ = roc_curve(y_true=true_y_bin[:, i], y_score=y_score[:, i])
        if any(numpy.isnan(tpr[i])): 
            roc_auc[i] = numpy.nan
        else:
            roc_auc[i] = auc(fpr[i], tpr[i])
        p_1 = numpy.mean(true_y_bin[:, i])   # prior probability of class i
        mean_error[i] = (1.-p_1)*fpr[i] + p_1*(1.-tpr[i])

    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(y_true=true_y_bin.ravel(), y_score=y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
    
    return (fpr, tpr, roc_auc, mean_error)
```

Implement a bucket-labeling function that aggregates labels of examples generated from that bucket:

```python
def step(x):
    return 1 * (x > 0)

def extract_bucket_label(bucket_id,buckets):   # find the label of a bucket bucket_id, assuming this bucket is in the buckets set passed as 'buckets'
    attr = [key for key, value in buckets.iteritems() if bucket_id in value]
    if len(attr) != 1:
        print 'WARNING: expect to find one label, found %i labels' %len(attr)
        return attr
    else:
        return attr[0]

# Label a bucket from 'buckets_set' with given 'bucket_id':
def predict_bucket_label(classifier, bucket_id, buckets_set, regenerate_examples=False, thresholds=None):
    '''
    The function evaluates the soft label and the hard label (obtained from the soft label using the thresholds) of a given bucket with 'bucket_id'
    
    classifier: the classifier object to be used for predicting the soft labels of the examples
    thresholds: per-class decision thresholds to be used when collapsing the obtained bucket soft label to bucket hard label
    bucket_id: id of the bucket to label
    buckets_set: the set of the buckets this bucket belongs to
    regenerate_examples: should the examples be regenerated from this bucket (True), or should the previously generated examples from this bucket be used for labeling (False)?
    '''
    soft_label = {k:0 for k in attributes}  # key=attribute, value=probability, initialize with probs=0
    
    bucket_attr = extract_bucket_label(bucket_id, buckets_set)
    if regenerate_examples:
        bucket = {}   
        bucket[bucket_attr] = {}
        bucket[bucket_attr][bucket_id] = buckets_set[bucket_attr][bucket_id]      # extract the bucket corresponding to 'bucket_id'    
        examples = generate_examples(bucket,hp,True)   # generate examples from this bucket
    else:
        if bucket_id in list(examples_train['bucket_id']):
            examples_set = examples_train
        elif bucket_id in list(examples_validate['bucket_id']):
            examples_set = examples_validate
        elif bucket_id in list(examples_test['bucket_id']):
            examples_set = examples_test
        else:
            print ('bucket_id %s is not found in either set of examples. Recommend to set regenerate_examples to True.')
            return
        
        examples = examples_set[examples_set['bucket_id']==bucket_id]     # extract the examples previously generated from this bucket, from examples_set
    
    if len(examples.index)==0:
        print ('predict_bucket_label: No examples to label, stopping.')
        return

    # predict soft labels (labels' posteriors) of examples: (note that the decision thresholds are not relevant here!)
    x = examples.iloc[:,3:].as_matrix()  # feature vectors of the examples
    y = examples.iloc[:,2].as_matrix()
    x_shared, y_shared = shared_dataset(x, y)
    x = x_shared.get_value()   # this is done for type compatibility with theano function's argument for mlp_predict and logistic_predict
    
    if str(classifier.__class__) == "<class 'sklearn.ensemble.forest.RandomForestClassifier'>":
        y_score = classifier.predict_proba(x)
        
    if str(classifier.__class__) == "<class 'SdA.SdA'>":
        y_score = mlp_predict(classifier, x, proba=True, verbose=False)
    
    if str(classifier.__class__) == "<class 'logistic_sgd.LogisticRegression'>":
        y_score = logistic_predict(classifier, x, proba=True, verbose=False)
    
    y_score_mean = numpy.mean(y_score, axis=0)  # take a mean of each column in y_score - i.e., mean soft label of all examples
    
    for attr in attributes:
        soft_label[attr] = y_score_mean[attributes_dict[attr]]
    
    # Evaluate the hard label of the bucket, based on the thresholds:
    if thresholds is None:
        thresholds = numpy.repeat(1./len(attributes), len(attributes))   # default thresholds = equal thresholds for all classes
    else:
        thresholds = numpy.array([v for k, v in thresholds.iteritems()])
#     hard_label = targets_dict[numpy.argmax(y_score_mean*(y_score_mean >= thresholds))]  # choose the label whose score is the largest among those over their relative thresholds
    hard_label = targets_dict[numpy.argmax(y_score_mean - thresholds)]  # choose the label whose threshold-offset score is the largest
    
    return (soft_label, hard_label, y_score_mean)

def per_class_errors(conf_matrix, verbose=False):
    # Calculate per-class bucket misclassification errors (due to FN, and due to both FN and FP) from the passed confusion matrix:
    per_class_fn_error = dict()
    per_class_fnfp_error = dict()
    cm_actual = list(conf_matrix.axes[1])
    cm_predicted = list(conf_matrix.axes[0])
    N = conf_matrix.sum().sum()   # total number of examples of all classes
    for attr in attributes:
        if attr in cm_actual:
            if attr in cm_predicted:
                TP = conf_matrix[attr][attr]  # true positives for this class=attr
                FP = sum(conf_matrix.loc[attr]) - conf_matrix[attr][attr]  # false positives for this class (predicted to be of this class, but actually are not)
            else:
                TP = 0  # this attr is not predicted at all
                FP = 0  # no positive, true or false, predictions at all for this class
                
            N_attr = sum(conf_matrix[attr])   # total number of buckets of this class
            per_class_fn_error[attr] = 1. - float(TP)/N_attr   # FN/N_attr = (N_attr - TP)/N_attr
            per_class_fnfp_error[attr] = float((N_attr-TP)+FP)/N   # error due to both FP and FN: (FN+FP)/N
            if verbose: 
                print ('\nattr = %s: TP=%i, FP=%i, N_attr=%i, FN/N_attr=%f, (FN+FP)/N=%f' \
                    %(attr, TP, FP, N_attr, per_class_fn_error[attr], per_class_fnfp_error[attr]))

        else:  # no examples of this class in conf_matrix
            per_class_fn_error[attr] = numpy.nan
            per_class_fnfp_error[attr] = numpy.nan
    per_class_fn_error["!AVERAGE_per_class_"] = numpy.nanmean(per_class_fn_error.values())
    per_class_fnfp_error["!AVERAGE_per_class_"] = numpy.nanmean(per_class_fnfp_error.values())
    return per_class_fn_error, per_class_fnfp_error

# Calculate classifier's per-class ROC, AUC, mean errors (due to FP and FN) for different decision thresholds, and the optimal thresholds, on a given collection of 'buckets'
def bucket_labeler_roc_auc(classifier, buckets, regenerate_examples=False):
    # Create a list of ids of the buckets in 'buckets'
    bucket_ids = [buckets[attr].keys() for attr in buckets.keys()]
    bucket_ids = numpy.unique([item for sublist in bucket_ids for item in sublist])
    
    labels_true = [extract_bucket_label(b_id, buckets) for b_id in bucket_ids]   # extract true labels of buckets in 'buckets'
    
    mean_error = dict()   #per-class mean error, in 'this-vs-other classes' manner
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    thresh = dict()
    optimal_thresholds = dict()  # optimal decision thresholds for each class
    
    true_y_bin = label_binarize(y=[attributes_dict[l] for l in labels_true], classes=targets_dict.keys())
    # Evaluate soft labels of buckets in bucket_ids:
    y_score = numpy.array([predict_bucket_label(classifier, b_id, buckets, regenerate_examples)[2] for b_id in bucket_ids])
    for i in targets_dict.keys():
        fpr[i], tpr[i], thresh[i] = roc_curve(y_true=true_y_bin[:, i], y_score=y_score[:, i])
        if any(numpy.isnan(tpr[i])): 
            roc_auc[i] = numpy.nan
        else:
            roc_auc[i] = auc(fpr[i], tpr[i])
        
        # Evaluate the mean errors (due to both false positives and false negatives) for each class at different thresholds:
        p_1 = numpy.mean(true_y_bin[:, i])   # prior probability of class i
        mean_error[i] = (1.-p_1)*fpr[i] + p_1*(1.-tpr[i])
        
#         # Find the optimal threshold for class i - that for which the mean_error[i] is the smallest:
#         opt_th_ind = numpy.argmin(mean_error[i])
#         optimal_thresholds[i] = thresh[opt_th_ind]   # optimal threshold
        
        # Find the optimal threshold for class i - that for which the point on the ROC is closest to the point (tpr=1, fpr=0)
        distances = (tpr[i] - 1.)**2 + fpr[i]**2
        opt_th_ind = numpy.argmin(distances)
        if any(numpy.isnan(distances)):
            optimal_thresholds[i] = 0
        else:
            optimal_thresholds[i] = thresh[i][opt_th_ind]   # optimal threshold
        
    return (fpr, tpr, roc_auc, mean_error, optimal_thresholds)
```

Perform clustering of examples, with automatic learning of the number of clusters

Determine the best number of clusters in KMeans, using the gap statistic:

```python
# Wk is the sum of normalized intra-cluster distances, Wk = sum(1/n_k D_k, k=1,...,K), where n_k is the number of data points in 
# cluster C_k, and D_k = sum(sum(||x_i - x_j||^2, x_j \in C_k), x_i \in C_k) = 2n_k sum(||x_i - \mu_k||^2, x_i \in C_k), 
# and \mu_k is the centroid of the k-th cluster C_k.
# https://datasciencelab.wordpress.com/2013/12/27/finding-the-k-in-k-means-clustering/ 
def Wk(mu, clusters):
    K = len(mu)
    return sum([numpy.linalg.norm(mu[i]-c)**2/(2*len(c)) \
               for i in range(K) for c in clusters[i]])
```

```python
def cluster_points(X, mu):  # Assign datapoints X to clusters with centroids given by mu:
    clusters  = {}
    for x in X:
        bestmukey = min([(i[0], numpy.linalg.norm(x-mu[i[0]])) \
                    for i in enumerate(mu)], key=lambda t:t[1])[0]
        try:
            clusters[bestmukey].append(x)
        except KeyError:
            clusters[bestmukey] = [x]
    return clusters  #returns a dictionary with keys being the cluster centroids, values being the lists of points in that cluster

def find_centers(X, K):  # find the centers of K clusters, and assign datapoints X to the clusters:
    estimator = KMeans(init='k-means++', n_clusters=K, n_init=10, precompute_distances=True)
    estimator.fit(X)
    mu = estimator.cluster_centers_
    clusters = cluster_points(X, mu)
    return (mu, clusters, estimator)
```

```python
# The gap statistic:
def bounding_box(X):
    xmin, xmax = min(X,key=lambda a:a[0])[0], max(X,key=lambda a:a[0])[0]
    ymin, ymax = min(X,key=lambda a:a[1])[1], max(X,key=lambda a:a[1])[1]
    return (xmin,xmax), (ymin,ymax)
 
def gap_statistic(X, ks=range(1,10)):
    (xmin,xmax), (ymin,ymax) = bounding_box(X)
    # Dispersion for real distribution
    Wks = numpy.zeros(len(ks))
    Wkbs = numpy.zeros(len(ks))
    sk = numpy.zeros(len(ks))
    for indk, k in enumerate(ks):
        mu, clusters, estimator = find_centers(X,k)
        Wks[indk] = numpy.log(Wk(mu, clusters))
        # Create B reference datasets sampled uniformly from the data bounding box:
        B = 10    #make it odd?
        BWkbs = numpy.zeros(B)
        for i in range(B):
            Xb = []
            for n in range(len(X)):
                Xb.append([random.uniform(xmin,xmax),
                          random.uniform(ymin,ymax)])
            Xb = numpy.array(Xb)
            mu, clusters, estimator = find_centers(Xb,k)
            BWkbs[i] = numpy.log(Wk(mu, clusters))
        Wkbs[indk] = sum(BWkbs)/B   # Expected value of log(W_k) over the reference (uniform) distribution
        sk[indk] = numpy.sqrt(sum((BWkbs-Wkbs[indk])**2)/B)
    sk = sk*numpy.sqrt(1+1/B)
    return(ks, Wks, Wkbs, sk)
```

Label the clusters according to the labeled examples (where labels may include 'UNKNOWN'):

```python
def label_clusters(X, Y, mu):
# label the clusters using labels Y of examples X, and return labels with their probabilities within each cluster, as a dictionary
    cluster_labels = {}
    for x, y in zip(X, Y):
        cl = min([(i[0], numpy.linalg.norm(x-mu[i[0]])) \
                        for i in enumerate(mu)], key=lambda t:t[1])[0]  # find which cluster the point x belongs to
        try: 
            cluster_labels[cl].append(targets_dict[y])   # append the label to this cluster
        except KeyError:
            cluster_labels[cl] = [targets_dict[y]]

    cluster_label_freq = {}
    for ind in cluster_labels.keys():
        labels_list = numpy.unique(cluster_labels[ind])
        for attr in labels_list:
            prob = float(cluster_labels[ind].count(attr))/len(cluster_labels[ind])   #probability of this label in the cluster
            try:
                cluster_label_freq[ind].append([attr, prob])
            except KeyError:
                cluster_label_freq[ind] = [[attr, prob]]
    
    return cluster_label_freq

def cluster_predict(classifier, cluster_labels, new_x, verbose=False):   #classifier = model, new_set_x is the new data for which to predict the y label
    """
    Use the passed cluster object 'classifier' and cluster labels to predict labels for new_x (which should be a numpy ndarray).
    """

    mu = classifier.cluster_centers_   # extract cluster centroids
    
    # predict:
    y_pred = []
    for x in new_x:
        cl = min([(i[0], numpy.linalg.norm(x-mu[i[0]])) \
                        for i in enumerate(mu)], key=lambda t:t[1])[0]  # find which cluster the point x belongs to
        labels = [l[0] for l in cluster_labels[cl]]
        probs = numpy.array([l[1] for l in cluster_labels[cl]])
        y_pred.append(attributes_dict[labels[probs.argmax()]])

    
    return numpy.array(y_pred)
```
